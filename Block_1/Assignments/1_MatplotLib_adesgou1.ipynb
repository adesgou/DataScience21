## Exercise 6
Calculate total sale data for last year for each product and show it using a Pie chart
Note: In Pie chart display Number of units sold per year for each product in percentage.

***Note:*** use ``data[item].sum()`` to compute sums over columns 


### The result should look something like this:
<img src="IMG/6.png" width=400>

```python
data['toothpaste'].sum()
```

```python
my_dictionary={}
for i in data:
    my_dictionary[i]=data[i].sum()

axes_dict=my_dictionary.copy()

del axes_dict['month_number']
del axes_dict['total_profit']
del axes_dict['total_units']

my_labels=[]
my_values=[]
for j in axes_dict:
    my_labels.append(j)

    my_values.append(axes_dict[j]/my_dictionary['total_units']*100)

my_labels = [y.capitalize() for y in my_labels]

#print(my_dictionary)
#print(axes_dict)
#print(my_labels)
#print(my_values)
```

```python
plt.figure()

axes=plt.axes()
axes.yaxis.set_major_locator(ticker.MultipleLocator(500))
axes.set_xticks(np.arange(0, 13, 1))
axes.set_title('Sales data');
axes.grid(linestyle='--')
plt.pie(my_values, labels=my_labels, autopct='%1.1f%%')

axes.legend(loc='center right', bbox_to_anchor=(0.75, -0.26, 0.5, 1))
```
